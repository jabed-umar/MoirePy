# Angle Calculation and Moire Lattice Construction 

The details about angle calculation can be found [here](https://jabed-umar.github.io/MoirePy/angle_calculation_process/). In this section, we will see how to construct the commensurate moir√© lattice based on the commensurate angles.

```python
##import the necesaary libraries
import numpy as np
import matplotlib.pyplot as plt
##import the lattice types (triagle, square, hexagonal..etc)
from moirepy import BilayerMoireLattice, TriangularLayer, SquareLayer, KagomeLayer, HexagonalLayer
```

### Triangular Moire Lattice

```python
## Make the lattice (Lattice type, a(ll1), b(ll2), m(ul1), n(ul2), x_no, y_no)---  
## don't put fractional values for x_no and y_no, which are number of moire unit cells
lattice_tri = BilayerMoireLattice(TriangularLayer, 3, 4, 4, 3, 1, 1) 
##upper layer is not translated and pbc condition true by default.----
```

Please refer to the [calculator](../../find_theta/) to find your suitable values for $a$, $b$, $m$, and $n$.


```python
#Let's plot the lattice
lattice_tri.plot_lattice()
```

### Square Moire Lattice

```python
lattice_sqr = BilayerMoireLattice(SquareLayer, 3, 4, 4, 3, 2, 2) 
## We increse the number of moire unit cells along moire lattice vectors to 2,2
```

```python
#Let's plot the lattice
lattice_sqr.plot_lattice()
```

```python
# Suupose we want to access the moire lattice vectors and unit cell area of the square lattice
mlv1 = np.array(lattice_sqr.mlv1)  # first moire lattice vector
mlv2 = np.array(lattice_sqr.mlv2)  # second moire lattice vector
print("First moire lattice vector:", mlv1)
print("Second moire lattice vector:", mlv2)
print("Moire lattice area:", np.linalg.norm(np.cross(mlv1, mlv2)))
```

### Hexagonal Lattice (Graphene)

```python
lattice_hex = BilayerMoireLattice(HexagonalLayer, 3, 4, 4, 3, 1, 1) 
```

```python
lattice_hex.plot_lattice()
```

```python
## Suppose we want to translate the upper layer by a vector (0,1) in the moire lattice coordinates
lattice_hex_translate = BilayerMoireLattice(HexagonalLayer, 3, 4, 4, 3, 2, 2, (0, 1))
```

```python
lattice_hex_translate.plot_lattice()
```

### Kagome Lattice

```python
lattice_kagome = BilayerMoireLattice(KagomeLayer, 3, 4, 4, 3, 1, 1)
lattice_kagome.plot_lattice()
```
